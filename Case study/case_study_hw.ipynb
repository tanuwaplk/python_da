#### Как критики относятся к спортивным играм?

H0: средняя оценка критиков спортивных игр >= 7

H1: средняя оценка критиков спортивных игр менее 7

```python
sport_score = data[data['Genre'] == 'Sports'].Critic_Score.dropna()
```

```python
from scipy import stats as st
alpha = 0.05
result = st.ttest_1samp(sport_score, 7)
print(result)

if (result.statistic < 0) & (result.pvalue / 2 < alpha):
    print('Отвергаем нулевую гипотезу, среднее меньше 75, критикам не нравятся спортивные игры')
else:
    print('Не отвергаем нулевую нулевую гипотезу, критикам нравятся спортивные игры')
```

#### Критикам нравятся больше игры на PC или на PS4?

H0: средняя оценка критиков игр на PC >= PS4

H1: средняя оценка критиков игр на PC < PS4 

```python
PC_score = data[data['Platform'] == 'PC'].Critic_Score.dropna()
PS4_score = data[data['Platform'] == 'PS4'].Critic_Score.dropna()
```

```python

alpha = 0.05
result = st.ttest_ind(PC_score, PS4_score)

if  (result.pvalue / 2 < alpha):
    print('Отвергаем нулевую гипотезу, критикам больше нравятся игры на PS4')
else:
    print('Не отвергаем нулевую гипотезу, критикам больше нравятся игры на PС')
```

```python
PC_score.mean()
```

```python
PS4_score.mean()
```

#### Критикам больше нравятся стрелялки или стратегии?

```python
shooter_score = data[data['Genre'] == 'Shooter'].Critic_Score.dropna()
strategy_score = data[data['Genre'] == 'Strategy'].Critic_Score.dropna()
```

H0: средняя оценка критиков игр на Shooter >= Strategy

H1: средняя оценка критиков игр на Shooter < Strategy 

```python

alpha = 0.05
result = st.ttest_ind(shooter_score, strategy_score)

if  (result.pvalue / 2 < alpha):
    print('Отвергаем нулевую гипотезу, критикам больше нравятся Стратегии')
else:
    print('Не отвергаем нулевую гипотезу, критикам больше нравятся Стратегии')
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/obulygin/pyda_homeworks/master/stat_case_study/spam.csv', error_bad_lines=False)
```

```python
df
```

```python
df["Message"] = df["Message"].str.lower()
```

```python
df["Message"] = df["Message"].str.replace('[\W_]+',' ')
```

```python
df
```

```python
from nltk.corpus import stopwords
stopwords_set = set(stopwords.words('english'))
```

```python
def delete_stop_words(row):
    words = row.Message.split()
    without_sw = []
    for word in words:
        if word not in stopwords_set:
            without_sw.append(word)
    return without_sw
```

```python
df['without_sw'] = df.apply(delete_stop_words, axis=1)
```

```python
df
```

```python
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
```

```python
def lemmatize(row):
    list_ = [wordnet_lemmatizer.lemmatize(word) for word in row.without_sw]
    str1 = " " 
    return str1.join(list_)
```

```python
df['lemmatized_text'] = df.apply(lemmatize, axis=1)
```

```python
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(df.lemmatized_text)
names = tfidf.get_feature_names()
tfidf_matrix = pd.DataFrame(tfidf_matrix.toarray(), columns=names)
```

```python
tfidf_matrix
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(tfidf_matrix, df.Category, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(random_state = 42)
```

```python
lr.fit(X_train, y_train)
```

```python
lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, lr.predict(X_test))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, lr.predict(X_test))
```
