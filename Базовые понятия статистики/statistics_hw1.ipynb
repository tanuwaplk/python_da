- Проверка на выбросы величины ректальных температур

```python
q1_rt = df.rectal_temp.quantile(0.25)
q3_rt = df.rectal_temp.quantile(0.75)
iqr_rt = q3_rt - q1_rt
lower_bound_rt = q1_rt - (1.5 * iqr_rt) 
upper_bound_rt = q3_rt + (1.5 * iqr_rt)
```

```python
outliers_rt = df[df.rectal_temp.between(lower_bound_rt, upper_bound_rt, inclusive=True)]
null_rt= df[df.rectal_temp.isnull()]
df_remove_outliers_rt = pd.concat([outliers_rt,null_rt])
df_remove_outliers_rt
```

В документации есть информация о связи между rectal temp. и extremities_temp. Поскольку ректальная температура является непрерывной величиной, была предпринята попытка классификации ректуальной температуры на 4 категории в зависимости от попадаения в квантиль для дальнейшего заполнения пропусков значений в столбце extremities_temp

```python
def classify_temp(row):
    if row.extremities_temp <= df.rectal_temp.quantile(0.25):
        return "1"
    elif row.extremities_temp <= df.rectal_temp.quantile(0.50):
        return '2'
    elif row.extremities_temp <= df.rectal_temp.quantile(0.75):
        return '3'
    else:
         return '4'
```

```python
df['temp_class'] = df.apply(classify_temp, axis=1)
```

Возможно существует связь между пульсом и периферическим пульсом. Поскольку пульс является непрерывной величиной, была предпринята попытка классификации рпульса на 4 категории в зависимости от попадаения в квантиль для дальнейшего заполнения пропусков значений в столбце peripheral_pulse

```python
def classify_pulse(row):
    if row.pulse <= df.pulse.quantile(0.25):
        return "1"
    elif row.pulse <= df.pulse.quantile(0.50):
        return '2'
    elif row.pulse <= df.pulse.quantile(0.75):
        return '3'
    else:
         return '4'
```

```python
df['pulse_class'] = df.apply(classify_pulse, axis=1)
```
