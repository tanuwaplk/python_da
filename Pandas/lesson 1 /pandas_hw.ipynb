### Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
data = pd.read_csv('power.csv')
data.head()
```

```python
data[ data['country'].str.contains('esto|latv|lit', case=False) ]['country'].unique()
```

```python
baltic_countries = data.query('year >= 2005 & year <= 2010 & quantity > 0 & country in ("Estonia", "Latvia", "Latvia") & category in (4, 12, 21)')
baltic_countries
```

```python
baltic_countries.quantity.sum()
```
